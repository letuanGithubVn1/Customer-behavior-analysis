### **2Ô∏è‚É£ Calculations**

Import Pandas, Matplotlib and Seaborn:

````python
# YOUR CODE
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
````

Then load the data by the following link:\
https://raw.githubusercontent.com/letuanGithubVn1/Customer-behavior-analysis/refs/heads/main/customer_data.csv

````python
# YOUR CODE
df = pd.read_csv('https://raw.githubusercontent.com/letuanGithubVn1/Customer-behavior-analysis/refs/heads/main/customer_data.csv')
````

Let's take a look at the given data first. Output the first 5 rows:

````python
df.head()
````

As we know, a correlation can be calculated only based on numeric columns. So, let's make a list of numeric columns:

````python
print(df.select_dtypes(include=[np.number]))
````

````python
# YOUR CODE
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
numeric_cols
````

Then calculate a correlation matrix using the method `.corr()`

> **üí° Don't forget...**
>
> You need to use only numeric columns!

````python
# YOUR CODE
customer_corr = df[numeric_cols].corr()
customer_corr
````

Finally, let's visualise the correlation matrix using Seaborn heatmap.

Set the title **"Customer Behavior Correlation Matrix"** to the chart.

````python
# YOUR CODE
plt.figure(figsize=(10, 8))
sns.heatmap(customer_corr, annot=True, cmap='coolwarm')
plt.title('Customer Behavior Correlation Matrix')
````
